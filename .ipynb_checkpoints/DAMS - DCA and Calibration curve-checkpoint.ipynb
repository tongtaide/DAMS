```python
import keras as K
from keras import layers
from keras.layers import Input,Dense,Activation,Dropout
from keras.models import Model
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from  tensorflow.keras.models import load_model
from keras.optimizers import SGD,Nadam
```

```python
def create_model(learn_rate=0.01, momentum=0):
    # create model
    model = Sequential()
    model.add(Dense(20, input_dim=13, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    optimizer = SGD(lr=learn_rate, momentum=momentum)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
# create model
model = KerasClassifier(build_fn=create_model, nb_epoch=100, batch_size=10, verbose=0)
# define the grid search parameters
learn_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
momentum = [0.0, 0.2, 0.4, 0.6, 0.8, 0.9]
param_grid = dict(learn_rate=learn_rate, momentum=momentum)
grid = GridSearchCV(estimator=model, param_grid=param_grid,scoring = 'roc_auc', n_jobs=-1)
grid_result = grid.fit(x_train_standard1,y_train,class_weight={0:1,1:4})
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
#参数寻优之batch_size、epochs
def create_model():
    # create model
    model = Sequential()
    model.add(Dense(20, input_dim=13, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    # Compile model
    optimizer = SGD(lr=0.1, momentum=0.8)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
# create model
model = KerasClassifier(build_fn=create_model, verbose=0)
# define the grid search parameters
batch_size = [10, 20,30,40,50]
epochs = [10, 50, 100]
param_grid = dict(batch_size=batch_size, nb_epoch=epochs)
grid = GridSearchCV(estimator=model, param_grid=param_grid,scoring = 'roc_auc', n_jobs=-1)
grid_result = grid.fit(x_train_standard1,y_train,class_weight={0:1,1:4})
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
# for params, mean_score, scores in grid_result.grid_scores_:
#     print("%f (%f) with: %r" % (scores.mean(), scores.std(), params))
```

```python
def create_model(activation='relu'):
    # create model
    model = Sequential()
    model.add(Dense(20, input_dim=13, init='uniform', activation=activation))
    model.add(Dense(1, init='uniform', activation='sigmoid'))
    # Compile model
    optimizer = SGD(lr=0.2, momentum=0.8)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)
model = KerasClassifier(build_fn=create_model, nb_epoch=10, batch_size=50, verbose=0)
# define the grid search parameters
activation = ['softmax', 'softplus', 'softsign', 'relu', 'tanh', 'sigmoid', 'hard_sigmoid', 'linear']
param_grid = dict(activation=activation)
grid = GridSearchCV(estimator=model, scoring = 'roc_auc',param_grid=param_grid, n_jobs=-1)
grid_result = grid.fit(x_train_standard1,y_train,class_weight={0:1,1:4})
# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
def ann():
    (layer0,layer1,layer2,layer3,layer4) = (13,20,15,4,1)
    model = K.models.Sequential()
    init = K.initializers.glorot_uniform(seed=0)
    #输入层
    model.add(Dense(output_dim=layer1, input_dim=layer0, activation='relu'))
    model.add(Dropout(0.1,seed = 0))

    model.add(Dense(output_dim=layer2, input_dim=layer1, activation='relu'))
    model.add(Dropout(0.1,seed = 0))

    model.add(Dense(output_dim=layer3, input_dim=layer2, activation='relu'))
    model.add(Dropout(0.1,seed = 0))

    # 输出层
    model.add(Dense(output_dim=layer4, activation='sigmoid'))  

    optimizer = SGD(lr=0.2, momentum=0.8)

    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

print("Starting training ")
model=ann()
    
model.fit(x_train_standard1,y_train, batch_size= 50, nb_epoch=10, shuffle='True', verbose=0,class_weight={0:1,1:4})

print("Training finished \n")
# print('epoch:',i+1)
#variables16_model.save('my_model28.h5')
```

```python
from sklearn.datasets import make_classification as mc
from sklearn.metrics import brier_score_loss # 导入布里尔分数
from sklearn.calibration import calibration_curve # 对概率类模型进行校准，方法是分箱
from sklearn.linear_model import LinearRegression
```

```python
name = ["LR","RFC","SVM",'XGB','DNN']
name1 = ["LR","RFC","SVM",'XGB']
```

```python
#fraction_of_positives trueproba, mean_predicted_value predproba.
plt.figure(figsize=(8,6),dpi=300)
plt.plot([0,1],[0,1],label = "Perfectly calibrated") # 绘制对角线，把（0，0），（1，1）连起来
for clf,name_ in zip([lr,rfc1,svm,xgb,stroke],name):
    
    
    if name_ == 'DNN':        
        proba_pos = clf.predict(x_test_standard1)
    else:                
        proba_pos = clf.predict_proba(x_test_standard1)[:, 1]
    #对只有0，1的标签值进行分箱后才能画图
    clf_score = brier_score_loss(y_test, proba_pos)
    fraction_of_positives, mean_predicted_value = calibration_curve(y_test, proba_pos, n_bins=3)
    model= LinearRegression()
    model.fit(mean_predicted_value.reshape(-1,1),fraction_of_positives.reshape(-1,1))
    print("%s:" % name_)
    print("\tBrier: %1.3f" % (clf_score),'slope:%1.3f'%model.coef_ ,'intercept:%1.3f'%model.intercept_)
    plt.plot(mean_predicted_value, fraction_of_positives, "s-",label="%s (%1.3f)" % (name_, clf_score))

plt.xlabel("Mean predicted value")
plt.ylabel("Fraction of positives")
plt.ylim([-0.05, 1.05])
plt.legend(loc="lower right")
plt.title('Calibration plots (reliability curve)')

plt.show()
```

```python
prob1 = lr.predict_proba(x_test_standard1)[:,1]
prob2 = rfc1.predict_proba(x_test_standard1)[:,1]
prob3 = svm.predict_proba(x_test_standard1)[:,1]
prob4 = xgb.predict_proba(x_test_standard1)[:,1]
prob5 = stroke.predict(x_test_standard1)
score1 = brier_score_loss(y_test,prob1,pos_label=1)
score2 = brier_score_loss(y_test,prob2,pos_label=1)
score3 = brier_score_loss(y_test,prob3,pos_label=1)
score4 = brier_score_loss(y_test,prob4,pos_label=1)
score5 = brier_score_loss(y_test,prob5,pos_label=1)
```

```python
#svm
prob3 = svm.predict_proba(x_test_standard2)[:,1]
score3 = brier_score_loss(y_test,prob3,pos_label=1)
y_means3,proba_means3 = calibration_curve(y_test,prob3,n_bins=3)
plt.figure(figsize=(8,6))
plt.plot([0,1],[0,1],linestyle='--',label = 'Perfect calibration')
plt.plot(proba_means3,y_means3,"s-",label = 'SVM(score =%1.3f)'%score3)
plt.xlabel("Mean Predcited probability")
plt.ylabel("True probability for MRS>1")
plt.title('calibration curve')
plt.legend(loc="lower right")
plt.show()
```
